### Deployment Steps
- 1) Import Module (contains all the packages)
- 2) Define an instance of the model class inside the module
    - Calling out the class has to be made with arguments - These are the attributes of that class defined within the constructor "__INIT__"  -- They can be named anything (when you are calling out the class). What matters is what the class inside the module, and the files inside the modules to be unpickled are named properly
    - model: the file containing the fine tuned finalized version of the logistic regression model
    - scaler: the file contains the statistical parameters (mean,std etc) needed to adjust the magnitude of all numbers we have in the data set
- 3) Call out the methods for that class and use it with the new made instance object of that class
    - Load and clean: will load and preprocess the new data like before
    - Predicted outputs: Feed the cleaned data into the model and deliver the output we want 
- 4) The output: The probability column comes from predict_proba which outlines the probabilites of each observation being 0 (first column) or 1 (second Column) but in the module we chose to show only the probability of being actually absent which the probability of being a 1 (predict_proba(:,1)) which is the second colum of the probability
