# Final Project - VIDEO GAMES SALES (1980-2017)

## Introduction

https://www.kaggle.com/gregorut/videogamesales

This dataset contains a list of video games with sales greater than 100,000 copies. It was generated by a scrape of vgchartz.com.

Fields include

Rank - Ranking of overall sales

Name - The games name

Platform - Platform of the games release (i.e. PC,PS4, etc.)

Year - Year of the game's release

Genre - Genre of the game

Publisher - Publisher of the game

NA_Sales - Sales in North America (in millions)

EU_Sales - Sales in Europe (in millions)

JP_Sales - Sales in Japan (in millions)

Other_Sales - Sales in the rest of the world (in millions)

Global_Sales - Total worldwide sales.

## A) BUSSINESS UNDERSTANDING

Stakeholder want to understand better the type of games that are most succesful considering, their Genre, Publishers, and Their Platform (Console)

### Hypothesis Formulation

+ __Null Hypothesis__: NA Sales ARE NOT affected by the Genre (lack of effect)
- __Alternative Hypothesis__: NA sales are affected by Genre (effect)

#### Discussing Releases (Games) Vs Sales (Global Millions) for each of the Genres, Platforms and Publishers

##### A) GENRE

```python
#__Genre Vs Releases (Copies)__


top_global_sales = no_Nans.groupby('Genre').agg('count').sort_values(by='Global_Sales',ascending=False)
plt.figure(figsize=(20,10))
sns.barplot(top_global_sales.index,top_global_sales['Global_Sales'],palette='plasma')
plt.title('Genre Vs Releases',size=20)
plt.xlabel('Genre',size= 15)
plt.ylabel('Total Release (Count of Games)',size= 15)



#__Genre Performance in Millions__


top_global_sales = no_Nans.groupby('Genre')[['Global_Sales']].agg('sum').sort_values(by='Global_Sales',ascending=False)


plt.figure(figsize=(20,10))
sns.barplot(top_global_sales.index,top_global_sales['Global_Sales'],palette='plasma')
plt.title('Genre Total Global Sales in Millions',size=20)
plt.xlabel('Genre',size= 15)
plt.ylabel('Total Global Sales (Millions)',size= 15)
```

__***We can see that the number of games released for Action and Sports Genre proportional with the Total Global sales, however for the Shooter Genre, even though it has less games, it resulted in more global sales landing it in the third place***__

#### B) PLATFORM

```python
#__Platform Vs Releases__

top_global_sales = no_Nans.groupby('Platform').agg('count').sort_values(by='Global_Sales',ascending=False)
plt.figure(figsize=(20,10))
sns.barplot(top_global_sales.index,top_global_sales['Global_Sales'])
plt.title('Platform VS Releases',size=20)
plt.xlabel('Platform Games (CONSOLE)',size= 15)
plt.ylabel('Total Release (Count of Games)',size= 15)


#__Platform Performance in Millions__


top_global_sales = no_Nans.groupby('Platform')[['Global_Sales']].agg('sum').sort_values(by='Global_Sales',ascending=False)
plt.figure(figsize=(20,10))
sns.barplot(top_global_sales.index,top_global_sales['Global_Sales'])
plt.title('Platform Total Global Sales in Millions',size=20)
plt.xlabel('Platform Games (CONSOLE)',size= 15)
plt.ylabel('Total Global Sales (Millions)',size= 15)

```

__***The Highest amount of games were released for DS and PS2 but when considering the Global Sales, PS2 Dominated the DS over the years. Also x360 games releases were less than DS games but landing the second place as the highest global sales***__

#### C) Top 15 Global PUBLISHERS

```python
#__Publisher Vs Releases__

top_publisher_names_15 = no_Nans.groupby('Publisher')[['Global_Sales']].agg('sum').sort_values(by='Global_Sales',ascending=False).head(15).index

# Create a filter for the highest 15
global_sales_top_publishers = no_Nans[no_Nans['Publisher'].isin(top_publisher_names_15)]

# Group by again to create the bar plot
global_sales_top_publishers = global_sales_top_publishers.groupby("Publisher").agg('count').sort_values(by='Global_Sales',ascending=False)
labels = global_sales_top_publishers.index

plt.figure(figsize=(15,10))
graph = sns.barplot(x=labels,y='Global_Sales',data=global_sales_top_publishers)
graph.set_xticklabels(labels=labels, rotation=90)
plt.title('Publisher Vs Releases',size=15)
plt.xlabel('Publisher',size= 15)
plt.ylabel('Total Release (Count of Games)',size= 15)


#__Top 15 Publishers Performance__ 


top_publisher_names_15 = no_Nans.groupby('Publisher')[['Global_Sales']].agg('sum').sort_values(by='Global_Sales',ascending=False).head(15).index

# Create a filter for the highest 15
global_sales_top_publishers = no_Nans[no_Nans['Publisher'].isin(top_publisher_names_15)]

# Group by again to create the bar plot
global_sales_top_publishers = global_sales_top_publishers.groupby("Publisher")[['Global_Sales']].agg('sum').sort_values(by='Global_Sales',ascending=False)
labels = global_sales_top_publishers.index

plt.figure(figsize=(15,10))
graph = sns.barplot(x=labels,y='Global_Sales',data=global_sales_top_publishers)
graph.set_xticklabels(labels=labels, rotation=90)
plt.title('Publisher Global Sales in Millions',size=15)
plt.xlabel('Publisher',size= 15)
plt.ylabel('Total Global Sales(Millions)',size= 15)
```

__***Electronic Arts Released the Highest number of games over the years, followed by Activision and Namco, However when Considering the Global Sales, NINTENDO has dominated the Market over Electronic Arts with over 1750 Million in Global sales from 1980-2016 with only 700 releases vs 1300 releases of EA Games***__

### Global Sales - Time Series - Top 5 Genres

#### __Top 5 Genres based on Global Sales__

```python
top_global_genres_names = no_Nans.groupby('Genre')['Global_Sales'].agg('sum').sort_values(ascending=False).head(5).index
top_genres = no_Nans[no_Nans['Genre'].isin(top_global_genres_names)]


fig, (ax0,ax1) = plt.subplots(2,2, figsize=(20,15))

fig.suptitle('Time Series Top 5 Best Global Genres , in Different Continents',size=20)
sns.lineplot(x='Year',y='NA_Sales',data=top_genres,hue='Genre',ax=ax0[0])
ax0[0].set_title('North American Sales based on Top 5 Global Genres',size=20)

sns.lineplot(x='Year',y='EU_Sales',data=top_genres,hue='Genre',ax=ax0[1])
ax0[1].set_title('European Sales based on Top 5 Global Genres',size=20)

sns.lineplot(x='Year',y='JP_Sales',data=top_genres,hue='Genre',ax=ax1[0])
ax1[0].set_title('Japanese Sales based on Top 5 Global Genres',size=20)

sns.lineplot(x='Year',y='Global_Sales',data=top_genres,hue='Genre',ax=ax1[1])
ax1[1].set_title('Global Sales based on Top 5 Global Genres',size=20)

plt.show()
```

## F) Data Transformation

Now to start the predictive model I will consider different categorical variables and converting them into their dummies (One hot encoded) and also I will remove very high outliers for the North American Sales

### Dealing with Outliers for NA sales - Since we want to predict NA sales

```python
sns.distplot(no_Nans['NA_Sales'])
```

#### Using the Interquartile Rule to Remove Outliers

```python
# Considering 95% of the data in NA Sales using 95% Percentile and removing all sales lower than 0 

#q = no_Nans['NA_Sales'].quantile(0.95)
#data_1 = no_Nans[no_Nans['NA_Sales']<q]

#data_1 = data_1[data_1['NA_Sales']>0]

#sns.distplot(data_1['NA_Sales'])
#data_1.shape


# 1) North American Sales: 
# A) Filtering using the IQR Rule

#Filtering using the IQR Rule
Q1=no_Nans['NA_Sales'].quantile(0.25)
Q3=no_Nans['NA_Sales'].quantile(0.75)
IQR = Q3-Q1

#If its not an outlier then its good otherwise NAN it
data_1 = no_Nans.where(~((no_Nans['NA_Sales']<Q1 - 1.5*IQR) | (no_Nans['NA_Sales']>Q3 + 1.5*IQR)))
data_1 = data_1.dropna()

# Now Considering only above 0 Millions North AMerican Sales

data_1 = data_1[data_1['NA_Sales']>0]
sns.distplot(data_1['NA_Sales'])
data_1.reset_index(drop=True,inplace=True)
data_1.shape
```

__Linear Regression Ordinary Least Square Assumptions__
- Linearity
- Normality and mean zero for residuals distribution
- Homoscedasticity (residual vs predicted are not sparse)
- No Endogenity (No correlation between features and error)
- No Autocorrelation (Happens in Time Series data, error terms should not be correlated and their variance is not constant but changing and random)
- No Multicollinearity - or no correlation between the features themselves (Variable Inflation factor can be used to check correlation between features)

### Lasso Regression With Cross Validation Using 3 Label Encoded Variables, and NA Sales as Target

__MODEL 6__


```python
y= df_5.iloc[:,-1]
x= df_5.iloc[:,3:-1]
```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score, cross_val_predict

lasso = Lasso()

grid = GridSearchCV(lasso, param_grid={'alpha': range(1, 10)}, refit=True)

y_pred_lasso = cross_val_predict(grid, x, y, cv=5)
```
